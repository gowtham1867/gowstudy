```python
concepts = [
    ['PG', 'Yes', 'Strong', 'Good'],
    ['UG', 'Yes', 'Strong', 'Good'],
    ['UG', 'No',  'Strong', 'Good'],
    ['PG', 'Yes', 'Weak',   'Average']
]

labels = ['Yes', 'Yes', 'No', 'No']

num_attributes = len(concepts[0])
S = ['0'] * num_attributes
G = [['?'] * num_attributes]

for i in range(len(concepts)):
    if labels[i] == 'Yes':
        for j in range(num_attributes):
            if S[j] == '0':
                S[j] = concepts[i][j]
            elif S[j] != concepts[i][j]:
                S[j] = '?'
        G = [g for g in G if all(g[j] == '?' or g[j] == S[j] for j in range(num_attributes))]
    else:
        new_G = []
        for g in G:
            for j in range(num_attributes):
                if g[j] == '?' and S[j] != concepts[i][j]:
                    h = g.copy()
                    h[j] = S[j]
                    new_G.append(h)
        G = new_G

    print("S =", S)
    print("G =", G)

print("Final S =", S)
print("Final G =", G)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

X = np.array([[800], [1000], [1200], [1500], [1800], [2000]])
y = np.array([40, 50, 60, 75, 85, 95])

model = LinearRegression()
model.fit(X, y)

y_pred = model.predict(X)

mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

n = len(y)
p = X.shape[1]

adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

print("MAE:", mae)
print("MSE:", mse)
print("R-Squared:", r2)
print("Adjusted R-Squared:", adj_r2)

```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, f1_score, roc_curve, auc

X = np.array([
    [5], [15], [25], [35], [45], [55], [65], [75]
])

y = np.array([0, 0, 0, 0, 1, 1, 1, 1])

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=1
)

model = LogisticRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:, 1]

cm = confusion_matrix(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

print("Confusion Matrix:\n", cm)
print("F1 Score:", f1)

plt.plot(fpr, tpr)
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve")
plt.show()
```
